```python
import pandas as pd
import numpy as np
```

```python
se2 = pd.read_csv('2016_se2_stat_features.csv')
se2.head()
```

```python
for i in se2.columns:
    print(i)
```

```python
LMS_features_merge = pd.DataFrame(se2['MASKED_STUDENT_ID'])
features = [
    'LOGIN_ATTEMPT', 'SESSION_TIMEOUT', 'LOGOUT', 'group', 'db',
    'myinfo', 'course', 'journal', 'email', 'staff',
    'annoucements', 'content', 'grade'
]

for f in features:
    LMS_features_merge[f] = se2['first_week'+ f] + \
                    se2['second_week'+ f] + \
                    se2['third_week'+ f] + \
                    se2['forth_week'+ f] + \
                    se2['fifth_week'+ f] + \
                    se2['sixth_week'+ f] + \
                    se2['seventh_week'+ f] + \
                    se2['eighth_week'+ f] + \
                    se2['nineth_week'+ f] + \
                    se2['tenth_week'+ f] + \
                    se2['eleventh_week'+ f] + \
                    se2['twelfth_week'+ f] + \
                    se2['thirteenth_week'+ f] + \
                    se2['forteenth_week'+ f] 
LMS_features_merge.head()
```

```python
Std_normal = pd.read_csv('Std_list_normal_2016_se2.csv')
Std_atrisk = pd.read_csv('Std_list_atRist_2016_se2.csv')
Std_all = pd.concat([Std_normal, Std_atrisk])
print(Std_all.shape)
Std_all.head()
```

```python
LMS_features_merge_mtx = pd.merge(Std_all, LMS_features_merge, on='MASKED_STUDENT_ID', how='left').fillna(0)
print(LMS_features_merge_mtx.shape)
print('')
```

```python
LMS_features_merge_mtx.to_csv('2016_se2_LMS_features_with_GPA.csv', index=False)
```
