# Demostration for At-risk Student Prediction

In this demo, we will illustrate how to use our model for **At-risk  (GPA lower than 2.0 in one semester) Student Prediction**.<br>
We use the following information:

- The clickstream on the Blackboard system (**LMS Record**)
- The library check-in records

Moreover, we also extract **multi-scale regularity** features from the above information <br> and the **social homophily** features from the library check-in records for prediction. <br>

In our previous investigation, we found there are 2%~4% of students to be at-risk among around 15000 students in previous results. <br>

The **input** of this demo is the above raw data of students in one semester. <br>
We will **output** the most probable **500** students as well as their probabilities to be at-risk.

## Raw Data Display

### Raw LMS records (example)

```python
df_test = pd.read_csv('DR0008_activity_accumulator_2016_09.csv', nrows=10, sep='\t')
df_test
```

<font size=4>**Data Description**</font>

- **De-id**: Refer to the De-identity id
- **Event_type**: The type of activity:
    - LOGIN_ATTEMPT: a user attempted to login to the system
    - LOGOUT: a user clicked the logout button
    - SESSION_INT: a user connected to the system and a session was created
    - PAGE_ACCESS: a page was accessed
    - COURSE_ACCESS: a page within a course was access
    - TAB_ACCESS: a tab was access
    - MODULE_ACCESS: triggered when a module was rendered
    - SESSION_TIMEOUT: a user’s session is involuntarily ended because no action was taken in a specific window
    - CONTENT_ACCESS: triggered when content is displayed
- **Course_pk1**: The key of the course record
- **Course_id**: For course activities; LMS course site id
- **Group_pk1**: Indicate the activities is group related / the reference id
- **Forum_pk1**: Indicate the forum ID in the discussion board
- **Internal_handle**: Brief reference name for LMS resources
    When using this field, the prefix helps identify the path the user took to access the page:
    - CP: indicates that the user went through the course control panel to view the page
    - PA: associates with portal administrator functions
    - ADMIN: refers to accesses through the administrator control panel
    - LIST: indicates that the page was requested after a search function
    - CUST: refers to a customization interface
- **Content_pk1**: Reference id for course content id
- **Data**: Description on the activity / the path of LMS resources / system message
- **Timestamp**: The date and time at which the event occurred
- **Session_id**: User activity session id
- **Mode**: This is the flag showing the mode of browsing, in mobile / web mode
    - When the value is “0”, the record is not yet updated;
    - When the value is “1”, the record is in mobile mode;
    - When the value is “2”, the record is in web mode;
