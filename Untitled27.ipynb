```python
df_se1_stat = pd.read_csv('2016_se1_stat_features.csv', index_col=0)
df_se1_stat.head()
```

```python
blacklist = ['workday','weekend','morning','afternoon','evening','overnight','workday_ExamMonth','weekend_ExamMonth','morning_ExamMonth','afternoon_ExamMonth','evening_ExamMonth','overnight_ExamMonth','workday_notExamMonth','weekend_notExamMonth','morning_notExamMonth','afternoon_notExamMonth','evening_notExamMonth','overnight_notExamMonth','workday_firstMonth','weekend_firstMonth','morning_firstMonth','afternoon_firstMonth','evening_firstMonth','overnight_firstMonth','examMonth','notExamMonth','firstMonth','total_checkin']
df_se1_stat_new = df_se1_stat[[i for i in df_se1_stat.columns if i not in blacklist]]
```

```python
df_se1_stat_new.shape
```

```python
def aggr(row, pre_fix):
    ans = 0
    for i in row.index:
        if i.endswith(pre_fix):
            ans += row[i]
    return ans


df_se1_stat_new['T_LOGIN_ATTEMPT'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='LOGIN_ATTEMPT')
df_se1_stat_new['T_SESSION_TIMEOUT'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='SESSION_TIMEOUT')
df_se1_stat_new['T_LOGOUT'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='LOGOUT')
df_se1_stat_new['T_group'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='group')
df_se1_stat_new['T_db'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='db')
df_se1_stat_new['T_myinfo'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='myinfo')
df_se1_stat_new['T_course'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='course')
df_se1_stat_new['T_journal'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='journal')
df_se1_stat_new['T_email'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='email')
df_se1_stat_new['T_staff'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='staff')
df_se1_stat_new['T_annoucements'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='annoucements')
df_se1_stat_new['T_content'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='content')
df_se1_stat_new['T_grade'] = df_se1_stat_new.apply(aggr ,axis=1, pre_fix='grade')
```
