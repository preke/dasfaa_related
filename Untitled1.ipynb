```python
import networkx as nx
from node2vec import Node2Vec
from IPython.display import display, Image
%matplotlib inline
# Create a graph
Oct_G = nx.Graph()
with open('lib_co_occ_Sept.edgelist', 'r') as f:
    Oct_G.add_weighted_edges_from([[int(j) for j in i.split()] for i in f.readlines() if int(i.split()[2]) > 380 ])
    
nx.draw(Oct_G, with_labels=False, edge_color='b', node_color='g', node_size=30, layout=nx.spring_layout(Oct_G, k=0.001,iterations=20))
# plt.show()
```

```python
node2vec_1 = Node2Vec(Oct_G, dimensions=64, walk_length=10, num_walks=50, workers=4) 
```
