```python
import pandas as pd
import numpy as np
```

```python
se2 = pd.read_csv('2016_se2_stat_features.csv')
se2.head()
```

```python
LMS_features_merge = pd.DataFrame(se2['MASKED_STUDENT_ID'])
features = [
    'LOGIN_ATTEMPT', 'SESSION_TIMEOUT', 'LOGOUT', 'group', 'db',
    'myinfo', 'course', 'journal', 'email', 'staff',
    'annoucements', 'content', 'grade'
]

for f in features:
    LMS_features_merge[f] = se2['first_week'+ f] + \
                    se2['second_week'+ f] + \
                    se2['third_week'+ f] + \
                    se2['forth_week'+ f] + \
                    se2['fifth_week'+ f] + \
                    se2['sixth_week'+ f] + \
                    se2['seventh_week'+ f] + \
                    se2['eighth_week'+ f] + \
                    se2['nineth_week'+ f] + \
                    se2['tenth_week'+ f] + \
                    se2['eleventh_week'+ f] + \
                    se2['twelfth_week'+ f] + \
                    se2['thirteenth_week'+ f] + \
                    se2['forteenth_week'+ f] 
LMS_features_merge.head()
```

```python
Std_normal = pd.read_csv('Std_list_normal_2016_se2.csv')
Std_atrisk = pd.read_csv('Std_list_atRist_2016_se2.csv')
Std_all = pd.concat([Std_normal, Std_atrisk])
print(Std_all.shape)
Std_all.head()
```

```python
LMS_features_merge_mtx = pd.merge(Std_all, LMS_features_merge, on='MASKED_STUDENT_ID', how='left').fillna(0)
print(LMS_features_merge_mtx.shape)
print('')
```

```python
LMS_features_merge_mtx.to_csv('2016_se2_LMS_features_with_GPA.csv', index=False)
```

```python
LMS_features_merge_mtx.head()
```

```python
Se1 = pd.read_csv('2016_se1_LMS_features_with_GPA.csv')
Se1.head()
```

```python
se1_se2 = pd.merge(LMS_features_merge_mtx, Se1, on = 'MASKED_STUDENT_ID', how='left', suffixes=['_se2', '_se1']).fillna(0)
se1_se2.shape
```

```python
se1_se2.head()
```

```python
se1_se2.columns
```

```python
se1_se2.to_csv('2016_2_sem_LMS_features', index=False)
```

```python
## Regression with GBDT
## with Se1 and Se2 features 
from sklearn.ensemble import GradientBoostingRegressor as GBR
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, accuracy_score, balanced_accuracy_score, recall_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

student_features = se1_se2 ## can be replaced with LMS_features_splited_month_mtx(the next block)

df_features = student_features[[i for i in student_features.columns if i != 'CUM_GPA_se2' and i != 'MASKED_STUDENT_ID']]
df_labels = pd.DataFrame(student_features['CUM_GPA_se2'])

X_train, X_test, y_train, y_test = \
    train_test_split(df_features, df_labels, test_size=0.2)

gbr = GBR()
gbr.fit(X_train, y_train)
gbr_preds = gbr.predict(X_test)
score = mean_squared_error(y_test, gbr_preds)
print(score)
```
