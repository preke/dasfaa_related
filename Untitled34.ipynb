```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G=nx.read_weighted_edgelist("lib_co_occ_se1_week_1_5.edgelist", nodetype=int) 
```

```python
pos = nx.circular_layout(G)


elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 10]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 10]


plt.figure(figsize=(20,15))
nx.draw_networkx_nodes(G, pos,node_size=10)
nx.draw_networkx_edges(G, pos, edgelist=elarge,
                       width=2,edge_color='r')
# nx.draw_networkx_edges(G, pos, edgelist=esmall,
#                        width=2, alpha=0.5, edge_color='r', style='dashed')
plt.axis('off')
plt.show()
```
